```python
from transformers import T5ForConditionalGeneration, T5Tokenizer

# Load model
model_name = "ramsrigouthamg/t5_paraphraser"

tokenizer = T5Tokenizer.from_pretrained(model_name)
model = T5ForConditionalGeneration.from_pretrained(model_name)

# Input sentence
sentence = "I am going to school"

# Format input for T5
text = "paraphrase: " + sentence + " </s>"

# Tokenize
encoding = tokenizer(text, return_tensors="pt")

# Generate paraphrases
outputs = model.generate(
    input_ids=encoding["input_ids"],
    attention_mask=encoding["attention_mask"],
    max_length=64,
    num_beams=10,
    num_return_sequences=5,

    # 🔥 anti-repetition controls
    repetition_penalty=2.5,
    no_repeat_ngram_size=3,

    # diversity control
    do_sample=True,
    top_k=50,
    top_p=0.95
)

# Decode results
print("\nGenerated Paraphrases:\n")
for i, out in enumerate(outputs):
    print(f"{i+1}. {tokenizer.decode(out, skip_special_tokens=True)}")
```

```python
from sentence_transformers import SentenceTransformer, util

model = SentenceTransformer("all-MiniLM-L6-v2")

original = "I like pizza"

candidates = [
    "I enjoy pizza",
    "Pizza is my favorite food",
    "I love pasta",
    "I like eating pizza"
]

emb = model.encode([original] + candidates, convert_to_tensor=True)

scores = util.cos_sim(emb[0], emb[1:])

best = sorted(zip(candidates, scores[0]), key=lambda x: x[1], reverse=True)

for s, score in best:
    print(score.item(), s)
```
